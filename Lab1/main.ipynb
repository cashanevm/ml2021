```python

arrayInt = np.arange(5)
print_array_info(arrayInt)

arrayInt = np.array([i for i in range(5, 10)])
print_array_info(arrayInt)

def randArray(itemsCount, min, max):
    return (max - min) * np.random.rand(itemsCount) - min


arrarndInt = randArray(10, 0, 10)
print_array_info(arrarndInt)

arrayInt = np.append(arrayInt, 11)
print_array_info(arrayInt)

for item in arrayInt:
    print(arrayInt)

arrayInt = np.insert(arrayInt, 6, 0)
print_array_info(arrayInt)

arrayInt = np.delete(arrayInt, 5)
print_array_info(arrayInt)

mtx = np.arange(16).reshape(4, 4)

print_array_info(mtx)

mtx = np.zeros((2, 2))

print_array_info(mtx)

mtx = np.ones((2, 2))

print_array_info(mtx)

mtx = np.arange(16, dtype=np.int32).reshape(4, 4)
other = randArray(16, 0, 5).reshape(4, 4)

print_array_info(mtx - other)
print_array_info(mtx + other)
print_array_info(mtx * other)
print_array_info(mtx ** other)
```

```python
def parabola2(x):
    return x ** 2

def parabola3(x):
    return x ** 3

def parabola05(x):
    return x ** 0.5

def tan(x):
    return np.tan(x)

def tanh(x):
    return np.tanh(x)

def sin(x):
    return np.sin(x)

def cos(x):
    return np.cos(x)

def linear(x):
    return -x

def linear2(x):
    return x

def linear0(x):
    return 0

def linear1(x):
    return 1

```

```python
def func(start, count, step, fn):
    x = []
    y = []
    stop = count * step + start + 1

    for i in np.arange(start, stop, step):
        x.append(i)
        y.append(fn(i))
    
    return np.array(x), np.array(y)
```

```python
figure, axis = plt.subplots()

x1, y1 = func(0, 100, 0.075, parabola2)
line1, = axis.plot(x1, y1, marker='s', color='orange')

x2, y2 = func(0, 100, 0.075, parabola3)
line2, = axis.plot(x2, y2, marker='^', color='green')

x3, y3 = func(0, 100, 0.075, parabola05)
line3, = axis.plot(x3, y3, marker='o',  color='red')

x0, y0 = func(0, 100, 0.075, lambda x: x / 10)
line0, = axis.plot(x0, y0, marker='x', color='blue')

axis.set_title('Some function graphics', fontsize=18)
axis.legend()
```

```python
def mae(actual, prediction):
    return np.absolute((actual - prediction))

def mse(actual, prediction):
    return np.square(actual - prediction)

def yhat(x, y):
    k, d = np.polyfit(x, y, 1)
    y_pred = k*x + d
    return y_pred
```

```python
parabola2Yhat = yhat(y1, y0)
parabola3Yhat = yhat(y2, y0)
parabola05Yhat = yhat(y3, y0)

print('mean parabola2(x):', parabola2Yhat.mean())

print('mean parabola3(x):', parabola3Yhat.mean())

print('mean parabola05(x):', parabola05Yhat.mean())

fig, ax = plt.subplots()

ax.plot(x1, parabola2Yhat)
ax.plot(x2, parabola3Yhat)
ax.plot(x3, parabola05Yhat)
```

```python

parabola2Mae = mae(y1, y0)
parabola3Mae = mae(y2, y0)
parabola05Mae = mae(y3, y0)

print('mean parabola2(x):', parabola2Mae.mean())

print('mean parabola3(x):', parabola3Mae.mean())

print('mean parabola05(x):', parabola05Mae.mean())




fig, ax = plt.subplots()

ax.plot(x1, parabola2Mae)
ax.plot(x2, parabola3Mae)
ax.plot(x3, parabola05Mae)
```

```python
parabola2Mse = mse(y1, y0)
parabola3Mse = mse(y2, y0)
parabola05Mse = mse(y3, y0)


print('mean parabola2(x):', parabola2Mse.mean())

print('mean parabola3(x):', parabola3Mse.mean())

print('mean parabola05(x):', parabola05Mse.mean())


fig, ax = plt.subplots()

ax.plot(x1, parabola2Mse)
ax.plot(x2, parabola3Mse)
ax.plot(x3, parabola05Mse)
```

```python
byParabola2 = (x1, y1, parabola2Yhat, parabola2Mae, parabola2Mse)
byParabola3 = (x2, y2, parabola3Yhat, parabola3Mae, parabola3Mse)
byParabola05 = (x3, y3, parabola05Yhat, parabola05Mae, parabola05Mse)


data = np.column_stack((byParabola2, byParabola3, byParabola05))
data = np.transpose(data, (1, 0))

graphdf = pd.DataFrame(data, columns=('X', 'Y', 'Y_hat', 'mAE', 'mSE'))

print(graphdf)
graphdf.to_csv('really_big_data.csv')
```
